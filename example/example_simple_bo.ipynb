```python
import torch

RANDOM_SEED = 47
torch.manual_seed(RANDOM_SEED)
```

```python
from BayesianOpt.problem import Rastrigin

prob = Rastrigin(ndim=1)
```

```python
from BayesianOpt.model import ExactGPModel

model = ExactGPModel(
    ndim=prob.ndim,
    lb=prob.lb,
    ub=prob.ub,
    interpolant=False,
)
```

```python
from BayesianOpt.design import QuasiMCDesign

design = QuasiMCDesign(
    ndim=prob.ndim,
    lb=prob.lb,
    ub=prob.ub,
    random_state=RANDOM_SEED,
)
```

```python
from BayesianOpt.acquisition import EI

acquisition = EI(
    ndim=prob.ndim,
    lb=prob.lb,
    ub=prob.ub,
)
```

```python
from BayesianOpt.base import SurrogateOptimization

algorithm = SurrogateOptimization(
    prob=prob,
    model=model,
    design=design,
    acquisition=acquisition,
)
```

```python
algorithm.run(
    T=10,
    plot_progress=True,
)
```
